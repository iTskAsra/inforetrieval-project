```python
## 6 points


import string
def clean_data(text : str):
    """Preprocesses the text with tokenization, case folding, stemming and lemmatization, and punctuations

    Parameters
    ----------
    text : str
        The title or abstract of an article

    Returns
    -------
    list
        A list of tokens
    """

    # TODO: tokenize, case_folding, stem, lemmatize, punctuations
    text_corrected = "".join([char for char in text if char not in string.punctuation])
    text_corrected = text_corrected.lower()
    tokens = nltk.word_tokenize(text_corrected)
    lemmatizer = nltk.stem.WordNetLemmatizer()
    stemmed = [lemmatizer.lemmatize(token) for token in tokens]
    return stemmed


data = clean_data("An abstract is a summary of the main article.") # return ["an", "abstract", "is", "a", "summary", "of", "the", "main", "article"]
print(data)
```
